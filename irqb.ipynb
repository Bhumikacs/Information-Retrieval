```python
# Implement an inverted index construction algorithm for the following two documents
# document1 = "The computer science students are appearing for a practical examination."
# document2 = "computer science practical examination will start tomorrow."
# Build a simple document retrieval system using the constructed index to find the documents containing terms “computer science”.

doc1 = "The computer science students are appearing for a practical examination."
doc2 = "computer science practical examination will start tomorrow."

docs = [doc1, doc2]
index = {}

# Create inverted index
for i, d in enumerate(docs, 1):
    for w in d.lower().replace(".", "").split():
        index.setdefault(w, set()).add(i)

# Search query
result = index["computer"] & index["science"]

print("Documents containing 'computer science':", result)
```

```python
# 1. Build a question-answering system using techniques such as information extraction.
# 2. Consider the following corpus:
# "India has the second-largest population in the world.",
# " It is surrounded by oceans from three sides which are Bay Of Bengal in the east, the Arabian Sea in the west and Indian oceans in the south.",
# "Tiger is the national animal of India.",
# "Peacock is the national bird of India.",
# "Mango is the national fruit of India."
# Build a question-answering system and query for "Which is the national bird of India?"

import spacy

nlp = spacy.load("en_core_web_sm")

# Corpus
sentences = [
"India has the second-largest population in the world.",
"It is surrounded by oceans from three sides which are Bay Of Bengal in the east, the Arabian Sea in the west and Indian oceans in the south.",
"Tiger is the national animal of India.",
"Peacock is the national bird of India.",
"Mango is the national fruit of India."
]
# Question
question = "Which is the national bird of India?"
# Extract keywords from question
keywords = {t.text.lower() for t in nlp(question) if not t.is_stop and not t.is_punct}
# Score sentences
scores = {}
for s in sentences:
    score = sum(1 for t in nlp(s.lower()) if t.text in keywords)
    scores[s] = score
# Find best matching sentence
answer = max(scores, key=scores.get)
print("Answer:", answer)
```

```python
# Implement a clustering algorithm (e.g., K-means or hierarchical clustering). Also, apply the clustering algorithm to a set of documents and evaluate the clustering results.
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Documents
docs = [
"Machine learning models require large datasets.",
"Deep learning uses complex neural networks.",
"Cloud computing provides scalable virtual servers.",
"AWS and Azure are popular cloud platforms."
]

# Convert documents to TF-IDF matrix
X = TfidfVectorizer(stop_words='english').fit_transform(docs)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=2, random_state=42)
labels = kmeans.fit_predict(X)

# Evaluate clustering
score = silhouette_score(X, labels)
print("Silhouette Score:", round(score,2))

# Show document clusters
for doc, label in zip(docs, labels):
    print("Cluster", label, ":", doc)
```

```python
#1.Implement a text classification algorithm
#2.Train the classifier on a labeled dataset and evaluate its performance.
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report

# Labeled dataset
data = {
    "text": ["high fever cough", "dry cough breathing issues",
             "happy joyful day", "sunny bright weather",
             "flu symptoms fever"],
    "label": ["Sick", "Sick", "Healthy", "Healthy", "Sick"]
}

df = pd.DataFrame(data)

# Split data into training and testing
X_train, X_test, y_train, y_test = train_test_split(
    df["text"], df["label"], test_size=0.2, random_state=42
)

# Convert text to numeric features
vectorizer = CountVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

# Train classifier
model = MultinomialNB()
model.fit(X_train_vec, y_train)

# Evaluate performance
pred = model.predict(X_test_vec)
print("Accuracy:", accuracy_score(y_test, pred))
print("\nClassification Report:\n", classification_report(y_test, pred))
```

```python
# Implement a learning to rank algorithm (e.g., RankSVM or RankBoost) on a given dataset.
import numpy as np
from sklearn.svm import LinearSVC

# Sample dataset (features of documents)
X = np.array([
    [1, 2],   # Doc1
    [2, 1],   # Doc2
    [3, 4],   # Doc3
    [4, 3]    # Doc4
])

# Relevance labels (higher = more relevant)
y = [2, 1, 2, 1]

# Train RankSVM model
model = LinearSVC()
model.fit(X, y)

# Predict ranking scores
scores = model.decision_function(X)

print("Ranking Scores:")
for i, s in enumerate(scores):
    print("Doc", i+1, ":", s)

# Rank documents based on scores
ranking = np.argsort(scores)[::-1]

print("Ranking order of documents:", ranking)
```

```python
# Consider a web graph with the following link structure:
# • Page A has links to pages B and C.
# • Page B has a link to page C.
# • Page C has links to pages
# Apply the PageRank algorithm and analyze the results.

import networkx as nx

# Create directed graph
G = nx.DiGraph()

# Link structure
G.add_edges_from([
    ('A','B'),   # A → B
    ('A','C'),   # A → C
    ('B','C')    # B → C
])

# Apply PageRank
pagerank = nx.pagerank(G, alpha=0.85)

# Display results
print("PageRank Scores:")
for page, score in pagerank.items():
    print(page, ":", round(score,4))
```

```python
# Implement cosine similarity to find similarity between following query and document. query="gold silver truck" document="shipment of gold damaged in a gold fire"
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Query and Document
query = "gold silver truck"
document = "shipment of gold damaged in a gold fire"

# Convert text into TF-IDF vectors
vectorizer = TfidfVectorizer()
tfidf = vectorizer.fit_transform([query, document])

# Compute cosine similarity
similarity = cosine_similarity(tfidf[0:1], tfidf[1:2])

print("Cosine Similarity:", similarity[0][0])
```

```python
# Handle challenges such as robots.txt, dynamic content, and crawling delays.
import requests
import time
from urllib.robotparser import RobotFileParser
from bs4 import BeautifulSoup

url =  "https://en.wikipedia.org/wiki/Information_retrieval"

# 1. Check robots.txt
rp = RobotFileParser()
rp.set_url(url + "/robots.txt")
rp.read()

if rp.can_fetch("*", url):

    # 2. Crawling delay (avoid overloading server)
    time.sleep(2)

    # 3. Fetch webpage
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    # Extract page text
    text = soup.get_text()
    print("Page Crawled Successfully")
    print(text[:300])

else:
    print("Crawling not allowed by robots.txt")
```

```python
# text summarization
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')

from nltk.tokenize import sent_tokenize, word_tokenize
from collections import Counter

text = """Natural language processing (NLP) is a field of computer science concerned with interactions between computers and human language.
Many challenges in NLP involve understanding and generating language.
Text summarization extracts important information to create a shorter version of text."""

sentences = sent_tokenize(text)

words = word_tokenize(text.lower())
freq = Counter(words)

scores = {s: sum(freq[w] for w in word_tokenize(s.lower())) for s in sentences}

print("Summary:", max(scores, key=scores.get))
```

```python
# Implement the Boolean retrieval model for the following corpus
# Document 1:BSc lectures start at 7.
# Document 2:My lectures are over.
# Document 3: Today is a holiday. Process the query “not lectures”

docs = {
1:"BSc lectures start at 7",
2:"My lectures are over",
3:"Today is a holiday"
}

query = "not lectures"
result = []

for i,t in docs.items():
    t = t.lower()
    if "not" in query:
        if query.split()[1] not in t:
            result.append(i)
    elif "or" in query:
        a,b = query.split(" or ")
        if a in t or b in t:
            result.append(i)
    elif "and" in query:
        a,b = query.split(" and ")
        if a in t and b in t:
            result.append(i)

print("Result:", result)
```

```python
# Write a program to calculate precision, recall and F-measure where true positive is 20, false positive is 10 and false negative is 30.
# Recall = Recall= TP /TP+FN
# Precision =Precision = TP / TP+FP
# F-score = 2 * (Precision * Recall) / (Precision + Recall)

# Given values
TP = 20
FP = 10
FN = 30

# Calculate metrics
precision = TP / (TP + FP)
recall = TP / (TP + FN)
f_score = 2 * (precision * recall) / (precision + recall)

print("Precision:", precision)
print("Recall:", recall)
print("F-measure:", f_score)
```

```python
# Given a set of predicted binary labels (y_pred) and their corresponding true binary labels (y_true), evaluate the performance by calculating precision, recall, f1-score and average precision of a binary classification model using standard metrics from scikit-learn module y_true = [0, 1, 1, 0, 1] y_scores = [0.1, 0.8, 0.6, 0.3, 0.9]

from sklearn.metrics import precision_score, recall_score, f1_score, average_precision_score

y_true = [0, 1, 1, 0, 1]
y_scores = [0.1, 0.8, 0.6, 0.3, 0.9]

# Convert scores to predicted labels
y_pred = [1 if s > 0.5 else 0 for s in y_scores]

# Calculate metrics
precision = precision_score(y_true, y_pred)
recall = recall_score(y_true, y_pred)
f1 = f1_score(y_true, y_pred)
avg_precision = average_precision_score(y_true, y_scores)

print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)
print("Average Precision:", avg_precision)
```
